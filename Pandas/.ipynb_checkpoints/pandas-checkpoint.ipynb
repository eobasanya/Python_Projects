```python
#Importing dependencies 
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import matplotlib as plt
import seaborn as sns
%matplotlib inline

```

```python
#read files and create df
whale_returns_csv = Path('whale_returns.csv')
algo_returns_csv = Path('algo_returns.csv')
sp500_returns_csv = Path('sp500_history.csv')
ALB_csv = Path('ALB_history.csv')
AVD_csv = Path('AVD_history.csv')
AVY_csv = Path('AVY_history.csv')
CBT_csv = Path('CBT_history.csv')
CCMP_csv = Path('CCMP_history.csv')
CTVA_csv = Path('CTVA_history.csv')
ENTG_csv = Path('ENTG_history.csv')
FMC_csv = Path('FMC_history.csv')
GWH_csv = Path('GWH_history.csv')
HXL_csv = Path('HXL_history.csv')
PLL_csv = Path('PLL_history.csv')
SQM_csv = Path('SQM_history.csv')






whale_returns_df = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns_df = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_returns_df = pd.read_csv(sp500_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
ALB_df = pd.read_csv(ALB_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
AVD_df = pd.read_csv(AVD_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
AVY_df = pd.read_csv(AVY_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
CBT_df = pd.read_csv(CBT_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
CCMP_df = pd.read_csv(CCMP_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
CTVA_df = pd.read_csv(CTVA_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
ENTG_df = pd.read_csv(ENTG_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
FMC_df = pd.read_csv(FMC_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
GWH_df = pd.read_csv(GWH_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
HXL_df = pd.read_csv(HXL_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
PLL_df = pd.read_csv(PLL_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
SQM_df = pd.read_csv(SQM_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)


```

```python
#clean data

#clean whale 
whale_returns_df.isnull().value_counts()
whale_returns_df.dropna(inplace=True)

#clean algo
algo_returns_df.isnull().value_counts()
algo_returns_df.dropna(inplace=True)

#clean sp500
sp500_returns_df.isnull().value_counts()
sp500_returns_df.dropna(inplace=True)

#remove sp500 $ sign
sp500_returns_df['Close'] = sp500_returns_df['Close'].str.replace('$', '')
sp500_returns_df['Close'] = sp500_returns_df['Close'].astype(float)


```

```python
lithium_stocks = pd.concat([ALB_df, AVD_df, AVY_df, CBT_df, CCMP_df, CTVA_df, ENTG_df, FMC_df, GWH_df, HXL_df, PLL_df, SQM_df], axis='columns', join='inner' )
lithium_stocks.sort_index(inplace=True)
lithium_stocks.columns = ['ALB', 'AVD', 'AVY', 'CBT', 'CCMP', 'CTVA', 'ENTG', 'FMC', 'GWH', 'HXL', 'PLL', 'SQM']
lithium_stocks
```

```python
#comparing by stock price
lithium_stocks.plot(figsize=(15,5))
```

```python
#Compare general statistical data
lithium_stocks.describe()
```

```python
#Comparing by percent change
lithium_daily_returns = lithium_stocks.pct_change()
lithium_daily_returns.plot(figsize=(15,5))
```

```python
#Comparing statistics by percent change
lithium_daily_returns.describe()
```

```python
weights = [0.083, 0.083,0.083,0.083,0.083,0.083,0.083,0.083,0.083,0.083,0.083,0.083]

lithium_total_returns_df = lithium_daily_returns.dot(weights)
lithium_total_returns_df.plot(figsize=(15,5))
```

```python
#lithium stocks cumulative returns
lithium_returns = (1 + lithium_total_returns_df).cumprod()
lithium_returns.plot(figsize=(15,5))
```

```python
#Hypothetical 10,000 investment
initial_investment = 10000
lithium_cumulative_profit = initial_investment * lithium_returns
lithium_cumulative_profit.plot(figsize=(15,5))
```

```python
lithium_cumulative_profit.describe()
```

```python
#Turn sp500 'Close' data to 'Returns'
sp500_returns_df = sp500_returns_df.pct_change()


#Reset index

sp500_returns_df.dropna()
```

```python
#joining dataframes
#First set indices. 
# sp500_returns_df.set_index(pd.to_datetime(sp500_returns_df['Date'], infer_datetime_format=True), inplace=True)
# sp500_returns_df.drop(columns=['Date'], inplace=True)

# whale_returns_df.set_index(pd.to_datetime(whale_returns_df['Date'], infer_datetime_format=True), inplace=True)
# whale_returns_df.drop(columns=['Date'], inplace=True)

# algo_returns_df.set_index(pd.to_datetime(algo_returns_df['Date'], infer_datetime_format=True), inplace=True)
# algo_returns_df.drop(columns=['Date'], inplace=True)

#algo_returns_df.head()
#Join dataframes
analysis_df = pd.concat([whale_returns_df, algo_returns_df, sp500_returns_df], axis=1)
analysis_df.rename(columns = {'Close': 'S&P 500'}, inplace=True)
analysis_df.dropna(inplace=True)
analysis_df
```

```python
analysis_cumulative_returns = (1 + analysis_df).cumprod() - 1
analysis_cumulative_returns.plot(figsize=(10,5), title='Daily Returns of All Portfolios')
```

```python
print(f'Soros Fund Management, Berkshire Hathaway, Inc., Tiger Global, and others outperform the S&P 500')
```

```python
#print boxplot for all of the returns
analysis_df.boxplot(figsize=(20,5))
```

```python
#Compare with lithium sector and determine most risky portfolios 
#Note: Unable to compare my lithium sector due to the time frame at which most of them went public. 
portfolios_statistics = analysis_df.describe()
analysis_stat_total = pd.concat([portfolios_statistics, lithium_total_returns_df.describe()], axis=1)
analysis_stat_total.columns = [*analysis_stat_total.columns[:-1], 'Lithium Sector']
analysis_stat_total
```

```python
analysis_df_std = analysis_df.std()
analysis_df_std.head()
```

```python
print('Berkshire Hathaway is riskier than the S&P 500 due to its standard deviation')
```

```python
#confirming risky portfolios by calculating beta
soros_covariance = analysis_df['SOROS FUND MANAGEMENT LLC'].cov(analysis_df['S&P 500'])
berkshire_covariance = analysis_df['BERKSHIRE HATHAWAY INC'].cov(analysis_df['S&P 500'])
tiger_covariance = analysis_df['TIGER GLOBAL MANAGEMENT LLC'].cov(analysis_df['S&P 500'])
#paulson_covariance = analysis_df['PAULSON & CO.INC.'].cov(analysis_df['S&P 500'])
algo1_covariance = analysis_df['Algo 1'].cov(analysis_df['S&P 500'])
algo2_covariance = analysis_df['Algo 2'].cov(analysis_df['S&P 500'])

#calculate variance of s&p 500
variance = analysis_df['S&P 500'].var()

#calculating betas
soros_beta = soros_covariance / variance
berkshire_beta = berkshire_covariance / variance
tiger_beta = tiger_covariance / variance
#paulson_beta = paulson_covariance / variance
algo1_beta = algo1_covariance / variance
algo2_beta = algo2_covariance / variance
print(f'Soros Beta: + {round(soros_beta, 10)} | Berkshire Beta: + {round(berkshire_beta, 10)} | Tiger Beta: + {round(tiger_beta, 10)} | Algo 1 Beta: + {round(algo1_beta, 10)} | Algo 2 Beta: + {round(algo2_beta, 10)}')
#analysis_df.head()
```

```python
#Calculate exponentially weighted moving average with 21 day window
analysis_cumulative_returns.rolling(window=21).mean().plot(figsize=(10,5), title='Cumulative Returns of All Portfolios')
```

```python
#Calculate Correlation 
price_correlation = analysis_df.corr()
#Plot Correlation
sns.heatmap(price_correlation, vmin=-2, vmax=2)
```

```python
#Calculate Sharpe Ratios
analysis_sharpe_ratios = (analysis_df.mean() * 252) / (analysis_df_std * np.sqrt(252))
analysis_sharpe_ratios
```

```python
#Plot sharpe ratios
analysis_sharpe_ratios.plot(kind='bar', title='Sharpe Ratios of all Portfolios')
```
